```python
mpg = sns.load_dataset("mpg")
mpg
```

```python
##Create a visual showing the relationship between horsepower and fuel economy.
sns.relplot(data=mpg, x="horsepower", y="mpg")
## the relationship between horsepower and fuel economy can be best described by the more fuel econonmy the less horspower.
## (Continuing) The less fuel economy the more horsepower.
```

```python
##Create a table summarizing fuel economy by country of origin
sns.catplot(data=mpg, x="origin", y="mpg", estimator="sum", kind="bar")
## the table underneath summarizes that usa by country has the highest fuel econonmy and europe has the least fuel economy.
##(continuing) This suggest that more driving takes place in the USA compared to Europe.
```

```python
##Create a visual showing how fuel economy has changed over time
sns.relplot(data=mpg, x="mpg", y="model_year")
## The relationship fuel economy has with time is that as the model years moves gets newer the more fuel econonmy it has.
##(Continued) The older the model year in relationship with time the less fuel econonmy is given to the model year.
```

```python
##Create two other visuals or tables related to fuel economy(1).
sns.relplot(data=mpg, x="mpg", y="weight")
## This visual relates to fuel economy by displaying the relationship for the weight of a vehicle with its fuel econonmy.
##(Continued) The more weight there is the less fuel econonmy is given.Less weight equates to more fuel economy.
```

```python
##Create two other visuals or tables related to fuel economy(2).
sns.catplot(data=mpg, x="cylinders", y="mpg", estimator="mean", kind="bar")
##The second visual relates to fuel economy by showing how many cylinders it takes for the mean fuel economy.
##(Continued) 4 cylinders is optimal for the most mean fuel economy with 5 cylinders being second.
```
